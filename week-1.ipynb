    Day 4-5: Lists

What are Lists?
-Lists are ordered collections that can hold multiple items. They're mutable (can be changed)

```python
# Creating lists
fruits = ["apple", "banana", "orange"]
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]

# Accessing elements
print(fruits[0])
print(fruits[-1])

# Modifying lists
fruits[1] = "mango"
fruits.append("grape")
fruits.insert(1, "kiwi")
fruits.remove("apple")
last = fruits.pop()

# List operations
print(len(fruits))
print("mango" in fruits)

# Slicing
numbers = [0, 1, 2, 3, 4, 5]
print(numbers[1:4])
print(numbers[:3])
print(numbers[3:])
```

    Day 6: Tuples

What are Tuples?
-Tuples are like lists but immutable (cannot be changed after creation). Use parentheses instead of brackets.

```python
# Creating tuples
coordinates = (10, 20)
colors = ("red", "green", "blue")
single = (5,)

# Accessing elements (same as lists)
print(coordinates[0])
print(colors[-1])

# Tuples are immutable
# coordinates[0] = 15       # This would cause an ERROR!

# Tuple unpacking
x, y = coordinates
print(x)
print(y)

# When to use tuples vs lists?
# - Tuples: Data that shouldn't change (coordinates, RGB values)
# - Lists: Data that will be modified (shopping cart, todo list)
```

    Day 7: Mini-Project - Simple Calculator

Project Requirements:
Build a calculator that can:

-Take two numbers as input from the user
-Ask which operation to perform (+, -, *, /)
-Display the result
-Handle division by zero

```python
# Simple Calculator Program

# Take two numbers as input from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Ask which operation to perform
print("\nSelect operation:")
print("1. Addition (+)")
print("2. Subtraction (-)")
print("3. Multiplication (*)")
print("4. Division (/)")

operation = input("Enter your choice (+, -, *, /): ")

# Perform the operation
if operation == '+':
    result = num1 + num2
    print(f"\nResult: {num1} + {num2} = {result}")

elif operation == '-':
    result = num1 - num2
    print(f"\nResult: {num1} - {num2} = {result}")

elif operation == '*':
    result = num1 * num2
    print(f"\nResult: {num1} Ã— {num2} = {result}")

elif operation == '/':
    if num2 == 0:
        print("\nError: Division by zero is not allowed.")
    else:
        result = num1 / num2
        print(f"\nResult: {num1} Ã· {num2} = {result}")

else:
    print("\nInvalid operation! Please enter one of +, -, *, or /.")
```

 Problem 3: Enhanced Calculator (Challenge)

-Adding support for power (** operator) and modulus (% operator)
-Validating that the operation entered is valid
-Allowing the user to perform multiple calculations without restarting
-Keeping a history of calculations in a list



```python

def add(a, b):
	return a + b

def subtract(a, b):
	return a - b 

def multiply(a, b):
	return a * b 

def divide(a, b):
	if b == 0:
		return 'Error: Division by zero!'
	return a / b 

def power(a, b):
	return a ** b 

def modulus(a, b):
	return a % b 


# Dictionary mapping - operators to functions
operations = {
	'+': add,
	'-': subtract,
	'*': multiply,
	'/': divide,
	'**': power,
	'%': modulus
}

def calculator():
	history = []
	print('Welcome to the calculator!')

	while True:
		try:
			num1 = float(input('\nEnter the first number: '))
			op = input('Enter an operation (+, -, *, /, **, %): ')

			# validate operator
			while op not in operations:
				print('Invalid operation, Please choose from given :')
				op = input('Enter a valid operation: ')

			num2 = float(input('Enter the second number: '))

			# perform calc
			result = operations[op](num1, num2)
			print(f'Result: {num1} {op} {num2} = {result}')

			# save to history
			history.append(f'{num1} {op} {num2} = {result}')
		except ValueError:
			print('Invalid input.!!')
			continue

		# ask user if they want another calc
		again = input('\nDo you want to perform another calc? (y/n): ')
		if again != 'y':
			print('\nCalc history: ')
			for entry in history:
				print(entry)
			print('\nThanks for using the calc.')
			break


# Run the calculator
if __name__ == '__main__':
	calculator()
```
