    Day 1-2: If/Else Statements

What are Conditional Statements?
-Conditionals allow your program to make decisions and execute different code based on conditions. They're like asking questions: "If this is true, do that; otherwise, do something else."

Comparison Operators:

```python
==    # Equal to
!=    # Not equal to
>     # Greater than
<     # Less than
>=    # Greater than or equal to
<=    # Less than or equal to
```

Logical Operators:

```python
and   # Both conditions must be True
or    # At least one condition must be True
not   # Reverses the boolean value
```

Basic If/Else Examples:

```python
# Simple if statement
age = 18
if age >= 18:
    print("You can vote!")
```

```python
# If/else
temperature = 25
if temperature > 30:
    print("It's hot!")
else:
    print("It's not too hot.")
```

```python
# If/elif/else (multiple conditions)
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(f"Your grade is: {grade}")
```

```python
# Logical operators
age = 25
has_license = True

if age >= 18 and has_license:
    print("You can drive!")
else:
    print("You cannot drive.")
```

```python
# Checking multiple conditions
weather = "rainy"
if weather == "sunny" or weather == "cloudy":
    print("Good day for a walk!")
else:
    print("Maybe stay inside.")
```

```python
# Not operator
is_weekend = False
if not is_weekend:
    print("Time to work!")
```

Nested If Statements:

```python
# If statements inside if statements
age = 20
has_ticket = True

if age >= 18:
    if has_ticket:
        print("Welcome to the concert!")
    else:
        print("You need a ticket.")
else:
    print("Sorry, you're too young.")
```

    Day 3-4: For Loops

What are For Loops?
-For loops let you repeat code a specific number of times or iterate through a collection (like a list).

Basic For Loop Examples:

```python
# Looping through a range
for i in range(5):
    print(i)
```

```python
for i in range(2, 7):
    print(i)
```

```python
# Range with step
for i in range(0, 10, 2):
    print(i)
```

```python
# Looping through a list
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
    print(f"I like {fruit}")
```

```python
# Looping through a string
word = "Python"
for letter in word:
    print(letter)
```

```python
# Using enumerate (get index and value)
colors = ["red", "green", "blue"]
for index, color in enumerate(colors):
    print(f"Color {index}: {color}")
```

```python
# Looping through a dictionary
person = {"name": "Alice", "age": 25, "city": "NYC"}
for key, value in person.items():
    print(f"{key}: {value}")
```

Practical Examples:

```python
# Sum of numbers
total = 0
for i in range(1, 6):
    total += i  # Same as: total = total + i
print(f"Sum: {total}")
```

```python
# Finding maximum
numbers = [34, 12, 89, 23, 56]
max_num = numbers[0]
for num in numbers:
    if num > max_num:
        max_num = num
print(f"Maximum: {max_num}")
```

```python
# Building a new list
original = [1, 2, 3, 4, 5]
doubled = []
for num in original:
    doubled.append(num * 2)
print(doubled)
```

    Day 5: While Loops

What are While Loops?
-While loops repeat code as long as a condition remains True. Use them when you don't know exactly how many times to loop.

Basic While Loop Examples:

```python
# Simple while loop
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# User input validation
password = ""
while password != "python123":
    password = input("Enter password: ")
print("Access granted!")
```

```python
# Countdown
num = 5
while num > 0:
    print(num)
    num -= 1
print("Blastoff!")
```

```python
# Accumulator pattern
total = 0
num = 1
while num <= 10:
    total += num
    num += 1
print(f"Sum of 1-10: {total}")
```

âš ï¸ Infinite Loops Warning:

```python
# This runs forever - DON'T DO THIS!
# while True:
#     print("This never stops!")

# Safe way to use while True
while True:
    answer = input("Continue? (yes/no): ")
    if answer.lower() == "no":
        break  # Exit the loop
```

    Day 6: Break and Continue

What are Break and Continue?

-break: Exits the loop immediately
-continue: Skips the rest of the current iteration and moves to the next

Break Examples:

```python
# Finding first even number
numbers = [1, 3, 5, 8, 9, 10]
for num in numbers:
    if num % 2 == 0:
        print(f"First even number: {num}")
        break
```

```python
# Search with break
names = ["Alice", "Bob", "Charlie", "David"]
search = "Charlie"
for name in names:
    if name == search:
        print(f"Found {search}!")
        break
else:  # This runs if loop completes without break
    print(f"{search} not found")
```

    Day 7: Nested Loops

What are Nested Loops?
-Loops inside loops. The inner loop completes all its iterations for each iteration of the outer loop.

Examples:

```python
# Simple nested loop
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
```

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}")
    print()
```

```python
# Pattern printing
for i in range(5):
    for j in range(i + 1):
        print("*", end="")
    print()
```
